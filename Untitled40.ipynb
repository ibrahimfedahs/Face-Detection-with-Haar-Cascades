```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# =====================================================
# 1. SAFE IMAGE LOADING FUNCTION
# =====================================================
def safe_read(path):
    img = cv2.imread(path, 0)
    if img is None:
        print(f"[WARNING] Could not load {path}. Using dummy image.")
        return np.zeros((200, 200), dtype=np.uint8)
    return img

# =====================================================
# 2. LOAD IMAGES SAFELY
# =====================================================
model = safe_read("img2.jpeg")
withglass = safe_read("img1.webp")
group = safe_read("img3.webp")

# =====================================================
# 3. LOAD HAAR CASCADES
# =====================================================
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# =====================================================
# 4. DISPLAY IMAGES
# =====================================================
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(model, cmap='gray')
plt.title('Model')

plt.subplot(1, 3, 2)
plt.imshow(withglass, cmap='gray')
plt.title('With Glasses')

plt.subplot(1, 3, 3)
plt.imshow(group, cmap='gray')
plt.title('Group')

plt.show()

# =====================================================
# 5. DETECTION FUNCTIONS
# =====================================================
def detect_face(img):
    face_img = img.copy()
    faces = face_cascade.detectMultiScale(face_img)
    for (x, y, w, h) in faces:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255,255,255), 2)
    return face_img

def adj_detect_face(img):
    face_img = img.copy()
    faces = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    for (x, y, w, h) in faces:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255,255,255), 2)
    return face_img

def detect_eyes(img):
    face_img = img.copy()
    eyes = eye_cascade.detectMultiScale(face_img)
    for (x, y, w, h) in eyes:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255,255,255), 2)
    return face_img

# =====================================================
# 6. APPLY FACE DETECTION
# =====================================================
result_glass = detect_face(withglass)
result_group = detect_face(group)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(result_glass, cmap='gray')
plt.title('Face Detection (Glasses)')

plt.subplot(1, 2, 2)
plt.imshow(result_group, cmap='gray')
plt.title('Face Detection (Group)')

plt.show()

# =====================================================
# 7. ADJUSTED FACE DETECTION
# =====================================================
result_adj_group = adj_detect_face(group)
plt.imshow(result_adj_group, cmap='gray')
plt.title('Adjusted Face Detection (Group)')
plt.show()

# =====================================================
# 8. EYE DETECTION
# =====================================================
result_eyes_model = detect_eyes(model)
result_eyes_glass = detect_eyes(withglass)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(result_eyes_model, cmap='gray')
plt.title('Eye Detection (Model)')

plt.subplot(1, 2, 2)
plt.imshow(result_eyes_glass, cmap='gray')
plt.title('Eye Detection (Glasses)')

plt.show()

# =====================================================
# 9. LIVE WEBCAM DETECTION
# =====================================================
print("\nStarting Live Webcam Face Detection...")

cap = cv2.VideoCapture(0)
plt.ion()
fig, ax = plt.subplots(figsize=(6, 6))

ret, frame = cap.read()
if not ret:
    print("Failed to open webcam.")
else:
    frame = detect_face(frame)
    im = ax.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.title('Live Video Face Detection')
    plt.show()

MAX_FRAMES = 300
frame_count = 0

while cap.isOpened() and frame_count < MAX_FRAMES:
    ret, frame = cap.read()
    if not ret:
        break

    frame = detect_face(frame)
    im.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.pause(0.01)

    frame_count += 1

cap.release()
plt.close(fig)

print("Webcam stream ended and resources released.")
```
